```python
import pandas as pd #import pandas
df= pd.read_csv('orders.csv') #readcsv file
df.head(20) #it reads top 20 file 
```

```python
df['Ship Mode'].unique() #check the unique values present in ship mode column
```

```python
df=pd.read_csv('orders.csv', na_values=['Not Available','unknown']) #we replaced 'not available' and 'unknown' data to null values here
df['Ship Mode'].unique() #show unique values in ship mode column 
```

```python
#change the column names 
df.columns=df.columns.str.lower() #converts column names to lower case
df
```

```python
df.columns = df.columns.str.replace(' ','_') #replace '-'(space) in colums with underscore(_)
df
```

```python
#derive new column discount, by calculating list price * discount price * 0.1
df['discount']=df['list_price']*df['discount_percent']*.01
df.head(5)
```

```python
#derive the new selling price column  by calculating list price - discount 
df['selling_price']=df['list_price'] - df['discount']
df.head()
```

```python
#derive new column profit by selling_price - cost_price
df['profit']= df['selling_price']-df['cost_price']
df.head()
```

```python
#convert order date from object data type to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d") #use capital Y as the date year has 4 digits 
df['order_date']
df.head()
```

```python
#drop cost price list price and discount perent columns
df.drop(columns=['cost_price','list_price','discount_percent'], inplace = True) 
df
```
